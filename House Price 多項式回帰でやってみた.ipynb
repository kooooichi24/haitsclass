```python
# 3つの説明変数のうちGrLivAreaとOverallQualを選択する。
# 散布図を見ると直線的な傾向よりも曲線的な傾向の方が見られる。
# 多項式回帰を選択する。
```

```python
# trainデータを用意
# DataFrameの.valuesメソッドでarrayに変換する
X = df_train.loc[:,["OverallQual","GrLivArea"]].values
X
```

```python
# trainデータを用意
# DataFrameの.valuesメソッドでarrayに変換する
y = df_train.loc[:,["SalePrice"]].values
y[:10]
```

```python
# 多項式回帰(2次関数を選択), GarageCarsを用いる
# 2次関数の変数
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)    # 2次の多項式規定を生成
X_quad = quad.fit_transform(X) # 生成した基底関数で変数変換を実行
X_quad
```

```python
# ホールド・アウト法によるデータの分割
# random_stateを0に固定してあるので､毎回同じサンプルに分割
from sklearn.model_selection import train_test_split
X_quad_train, X_quad_test, y_train, y_test = train_test_split(X_quad, y, test_size = 0.3, random_state = 0)
```

```python
# 線形回帰のインスタンスを生成
# 線形回帰のモデルを学習させる
from sklearn.linear_model import LinearRegression
model_quad = LinearRegression()

# 変換後の変数を利用して実行
model_quad.fit(X_quad_train, y_train)
```

```python
## 多項式回帰 3次関数を行う
```

```python
# 3次関数の変数変換
cubic = PolynomialFeatures(degree=3)     # 3次の多項式規定を生成
X_cubic = cubic.fit_transform(X) # 生成した基底関数で変数変換を実行
X_cubic
```

```python
# ホールド・アウト法による分割
X_cubic_train, X_cubic_test, y_train, y_test = train_test_split(X_cubic, y, test_size = 0.3, random_state = 0)
```

```python
# 線形回帰のインスタンスを生成
model_cubic = LinearRegression()

# 変換後の変数を利用して実行
model_cubic.fit(X_cubic_train, y_train)
```

```python
## 自由度調整済み決定係数の比較を行う
```

```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
# 自由度調整済み決定係数をtrainとtestに分けて出力
print('adjusted R^2')
print('')

# 2次関数
print('model_quad')
print('train: %.3f' % adjusted(model_quad.score(X_quad_train, y_train), len(y_train), 2))
print('test : %.3f' % adjusted(model_quad.score(X_quad_test, y_test), len(y_test), 2))
print('')

# 3次関数
print('model_cubic')
print('train: %.3f' % adjusted(model_cubic.score(X_cubic_train, y_train), len(y_train), 3))
print('test : %.3f' % adjusted(model_cubic.score(X_cubic_test, y_test), len(y_test), 3))
```

```python
## 2次関数と3次関数の間に差がないことが判明
## 70%を超えたので最適な説明変数はGrLivAreaとOverallQual、最適な回帰は多項式回帰とする。
```

```python
# 残差プロットを行う関数を定義
def res_plot(y_train, y_train_pred, y_test, y_test_pred):
    # train, testごとに残差を算出
    res_train = y_train_pred - y_train
    res_test  = y_test_pred - y_test
    
    # train, testごとに(予測値, 残差)をプロット
    plt.figure(figsize=(8, 8))
    plt.scatter(y_train_pred, res_train, color='blue', marker='o', label='train', alpha=0.5)
    plt.scatter(y_test_pred, res_test, color='green', marker='s', label='test', alpha=0.5)
    
    plt.xlabel('Predicted Values')                  # x軸のラベルを表示
    plt.ylabel('Residuals')                         # y軸のラベルを表示
    plt.legend(loc='upper left')                    # 凡例を左上に表示
    plt.hlines(y=0, xmin=-200000, xmax=700000, color='red') # 残差0の線を表示
    plt.xlim([-200000, 700000])                             # 表示する横軸の範囲を指定
    plt.show()
```

```python
# 残差プロットを出力
# 多項式回帰（3次関数の場合）
res_plot(y_train, model_cubic.predict(X_cubic_train), y_test, model_cubic.predict(X_cubic_test))
```
