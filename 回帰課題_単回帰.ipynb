```python
# csvファイルの読み込み
import pandas as pd
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# 正しく読み込まれたか確認
df_train.head()
```

```python
# データの詳細を出力
df_train.describe()
```

```python
# 相関係数行列を確認
df_train.corr()
```

```python
# DataFrameから1stFlrSF(１階)の列だけを取り出す
# locメソッドで列を指定できます
df_train.loc[:, ['1stFlrSF']].head()
```

```python
# 相関係数が一番大きかった1stFlrSFを用いる
# 1stFlrSFでSalePriceを予測してみる

# 1stFlrSFのデータをarrayに変換
X = np.array(df_train.loc[:, ['1stFlrSF']])
```

```python
# 変換されたことを確認
# 長いので､初めの5つだけ表示
X[:5]
```

```python
# SalePriceのデータをarray型に変換
y = np.array(df_train.loc[:, ['SalePrice']])
```

```python
# 変換されたことを確認
# 長いので､初めの5つだけ表示
y[:5]
```

```python
# グラフで全体的な傾向を捉える
# 散布図を出力
# 直線的な関係が確認できる

plt.scatter(X, y)                   # 散布図を表示
plt.title('1stFlrSF vs SalePrice')  # タイトルを表示
plt.xlabel('1stFlrSF')              # x軸のラベルを表示
plt.ylabel('SalePrice')             # y軸のラベルを表示
plt.show()                          # 指定した内容でグラフを出力

# これは直線的か？？
# 直線的ではないが単回帰で予測してみる
```

```python
# 線形回帰モデルのクラスを読み込み
from sklearn.linear_model import LinearRegression

# 線形回帰のインスタンスを生成
lr = LinearRegression()
```

```python
from sklearn.model_selection import learning_curve
lc = learning_curve(lr, X, y)
```

```python
# 単回帰モデルを学習
lr.fit(X, y)
```

```python
# 定数項を確認
lr.intercept_
```

```python
# 係数を確認
lr.coef_
```

```python
# 散布図に近似直線を重ねてみる
plt.scatter(X, y)                       # 散布図を表示
plt.plot(X, lr.predict(X), color='red') # 回帰直線を表示
plt.title('1stFlrSF vs SalePrice')     # グラフのタイトルを表示
plt.xlabel('1stFlrSF')               # x軸のラベルを表示
plt.ylabel('SalePrice')                  # y軸のラベルを表示
plt.show()                              # 上記の内容でグラフを表示
```

```python
## 予測してみる
# 1stFlrSFを入力する
test_1stFlrSF = 2300

# SalePriceを1stFlrSFから予測
n_pred = lr.predict(test_1stFlrSF)
print(n_pred)
```
